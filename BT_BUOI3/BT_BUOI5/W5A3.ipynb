```python
import math
def is_prime(a) -> bool:
    if a < 2:
        return False
    if a == 2 or a == 3:
        return True
    if a % 2 == 0 or a % 3 == 0:
        return False
    i = 5
    limit = math.isqrt(a)
    while i <= limit:
        if a % i == 0 or a % (i+2) == 0:
            return False
        i += 6
    return True
n = int(input())
print(is_prime(n))
```
