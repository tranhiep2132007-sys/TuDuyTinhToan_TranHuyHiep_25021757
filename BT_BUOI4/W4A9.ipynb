```python
def test(n) -> bool:
    s = str(n)
    for i in range(len(s)-1):
        for j in range(i+1,len(s)):
            if s[i] == s[j]:
                return False
    return True
import math
n = int(input())
for i in range(1,n+1):
    if math.sqrt(i) == round(math.sqrt(i)) and test(i) == True:
        print(i,end = " ")
```
