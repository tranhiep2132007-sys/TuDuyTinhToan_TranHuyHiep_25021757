```python
a = int(input())
def palindrome(n) -> bool:
    s = str(n)
    d = 0
    for i in range(round(len(s)/2)):
        if s[i] != s[len(s)-1-i]:
            return False
            d = 1
            break
    if d == 1:
        return False
    else:
        return True
d = 0
def reverse(n) -> int:
    s = str(n)
    k = ""
    for i in range(len(s)-1,-1,-1):
        k = k + s[i]
    n = int(k)
    return n
while palindrome(a) == False:
    a = a + reverse(a)
    d += 1
print(d,' ',a)

```
