```python
n = int(input())
import math
def is_prime(n) -> bool:
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    limit = math.isqrt(n)
    i = 5
    while i <= limit:
        if n % i == 0 or n % (i+2) == 0:
            return False
        i += 6
    return True
if is_prime(n) == True:
    print(n)
else:
    for i in range(round(n/2),1,-1):
        if is_prime(i) == True and n % i == 0:
            print(i)
            break

```
