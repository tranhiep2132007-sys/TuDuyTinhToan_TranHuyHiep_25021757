```python
import math
def is_prime(n) -> bool:
    if n < 2:
        return False
    if n in (2, 3):
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    limit = math.isqrt(n)  
    i = 5
    while i <= limit:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
t = 0
a, b = map(int, input().split())
for i in range(a,b+1):
    if is_prime(i) == True:
        t = t + i
print(t)
```
