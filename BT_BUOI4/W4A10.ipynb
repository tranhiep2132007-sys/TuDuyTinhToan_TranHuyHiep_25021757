```python
def collatz(n) -> int:
    d = 1
    while n > 1:
        if n % 2 == 0:
            n = n / 2
        else: 
            n = 3 * n + 1
        d = d + 1
    return d
n = int(input())
l = 0
a = 0
for i in range (1,n+1):
    if l < collatz(i):
        l = collatz(i)
        a = i
print(a," ",l)

        
```
