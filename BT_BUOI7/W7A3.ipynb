```python
def bubble_sort(arr):
    n = len(arr) 
    for i in range(n):
        swapped = False
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                swapped = True
                arr[j], arr[j+1] = arr[j+1], arr[j]
        if not swapped:
            break
    return arr
a = list(map(int, input().split()))
print(bubble_sort(a))
```
