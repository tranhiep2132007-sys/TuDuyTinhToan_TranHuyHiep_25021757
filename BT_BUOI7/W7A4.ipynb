```python
def most_frequent(arr):
    if not arr: return None
    count_dict = {}
    for x in arr:
        count_dict[x] = count_dict.get(x, 0) + 1
    
    max_count = 0
    res = arr[0]
    seen = set()
    for x in arr:
        if x not in seen:
            if count_dict[x] > max_count:
                max_count = count_dict[x]
                res = x
            seen.add(x)
    return res
arr = list(map(int, input().split()))
print(most_frequent(arr))
```
