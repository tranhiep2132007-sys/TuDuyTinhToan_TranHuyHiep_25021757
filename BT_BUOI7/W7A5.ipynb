```python
def count_pairs_with_sum(arr, X):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                count += 1
    return count
arr = list(map(int, input().split()))
X = int(input())
print(count_pairs_with_sum(arr, X))
```
