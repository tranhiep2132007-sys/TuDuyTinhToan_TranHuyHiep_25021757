```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def avg(self):
        if not self.scores: return 0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8.0: return "Excellent"
        elif 6.5 <= a < 8.0: return "Good"
        elif 5.0 <= a < 6.5: return "Average"
        else: return "Poor"
name = input().strip()
n = int(input())
stu = Student(name)
for _ in range(n):
    line = input().split()
    score = float(line[1])
    stu.add_score(score)
print(f"{stu.name} {stu.avg():.2f} {stu.rank()}")
```
