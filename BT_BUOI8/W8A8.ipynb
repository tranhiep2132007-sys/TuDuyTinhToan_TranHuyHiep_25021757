```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
line1 = input().split()
owner = line1[0]
balance = float(line1[1])
acc = BankAccount(owner, balance)
    
n = int(input())
for _ in range(n):
    cmd = input().split()
    action = cmd[0]
    val = float(cmd[1])
    if action == "DEPOSIT":
        acc.deposit(val)
    elif action == "WITHDRAW":
        acc.withdraw(val)
print(f"{acc.balance:.2f}")
```
