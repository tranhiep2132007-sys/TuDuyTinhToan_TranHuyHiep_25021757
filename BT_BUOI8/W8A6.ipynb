```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k
try:
    w, h, k = map(float, input().split())
    rect = Rectangle(w, h)
    rect.scale(k)
    print(f"{rect.area():.2f} {rect.perimeter():.2f}")
except:
    pass
```
